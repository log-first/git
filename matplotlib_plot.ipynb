```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
matplotlib.__version__
```

```python
# 数据
t = np.arange(0.0, 2.0, 0.01)
y = np.sin(2 * np.pi * t) + 1
```

```python
# 绘图
fig, ax = plt.subplots()
ax.plot(t, y) # 加载数据， 
ax.set(xlabel='time (s)', ylabel='voltage (mv)', title='About as simple at it get, folks') # 相关设置
ax.grid() # 显示格子
plt.show() # 显示图片
plt.savefig('test.png') # 保存图片
```

```python
# Scatter Demo2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
```

```python
with cbook.get_sample_data('goog.npz') as datafile:
    price_data = np.load(datafile)['price_data'].view(np.recarray)
price_data = price_data[-250:] # 获得 250 个交易日
deltal = np.diff(price_data.adj_close)/price_data.adj_close[: -1]
price_data
```

```python
# 标记大小，
volume = (15 * price_data.volume[:-2]/price_data.volume[0])**2
close = 0.003 * price_data.close[:-2]/0.003 * price_data.open[:-2]
print('volume = ', volume, 'close = ', close)
```

```python
# 绘图
fig, ax = plt.subplots()
ax.scatter(deltal[:-1], deltal[1:], c=close, s=volume, alpha=0.5)
ax.set_xlabel(r'$\Delta_i$', fontsize=15)
ax.set_ylabel(r'$\Delta_{i+1}$', fontsize=15)
ax.set_title('volume and percent change')

ax.grid(True)
fig.tight_layout()
plt.show()
```

```python
np.random.seed(19680801)
x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x**2 + y**2)
x, y, z
```

```python
plt.subplot(321)
plt.scatter(x, y, s=80, c=z, marker=">")

plt.subplot(322)
plt.scatter(x, y, s=80, c=z, marker=(6, 0))

verts = np.array([[-1, -1], [1, -1], [1, 1], [-1, -1]])
plt.subplot(323)
plt.scatter(x, y, s=80, c=z, marker=verts)

plt.subplot(324)
plt.scatter(x, y, s=80, c=z, marker=(4, 1))

plt.subplot(325)
plt.scatter(x, y, s=80, c=z, marker='+')

plt.subplot(326)
plt.scatter(x, y, s=80, c=z, marker=(5, 2))

plt.show()
```
